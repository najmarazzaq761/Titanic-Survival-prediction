```python
df.drop("Name",axis=1,inplace=True)
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="Pclass",y="Age",data=df,palette="GnBu_d").set_title("Age by passenger class")
plt.show()
```

```python
# add medians in missing data
def impute_age(cols):
  Age=cols[0]
  Pclass=cols[1]
  if pd.isnull(Age):
    if Pclass==1:
      return 37
    elif Pclass==2:
        return 29
    else:
      return 24
  else:
    return Age

# Apply the function to the Age column
df["Age"] =df[["Age","Pclass"]].apply(impute_age,axis=1)
```

```python
df.info()
```

```python
df.drop("PassengerId",axis=1,inplace=True)
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
objcat=["Sex","Embarked"]
for colname in objcat:
  df[colname]=df[colname].astype("category")
```

```python
df.select_dtypes(["category"]).columns
```

```python
sex=pd.get_dummies(df["Sex"],drop_first=True)
embarked=pd.get_dummies(df["Embarked"],drop_first=True)
df=pd.concat([df,embarked,sex],axis=1)
df.head(5)
```

```python
df.drop(["Sex","Embarked"],axis=1,inplace=True)
df.head(5)
```

```python
x=df.drop('Survived', axis=1)
y=df["Survived"]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(x_train,y_train)
```

```python
y_pred_logreg=logreg.predict(x_test)
print(y_pred_logreg)
print(y_pred_logreg.shape)
```

```python
probabilities=logreg.predict_proba(x_test)[:,1]
print(probabilities.shape)
print(probabilities)
```

```python
for i in range(len(y_test)):
  print(round(probabilities[i],3),"--->",y_pred_logreg[i])
```

```python
from sklearn.metrics import classification_report, accuracy_score
print("classification model")
print("-----"*40)
logreg_accuracy=round(accuracy_score(y_test,y_pred_logreg)*100,2)
print("accuracy is",logreg_accuracy,"%")
```

```python
df=pd.read_csv("/content/titanic_cleaned.csv")
df.head(5)
```

```python
x=df.drop("Survived",axis=1)
y=df["Survived"]
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn import preprocessing
pre_process=preprocessing.StandardScaler().fit(x)
x_transform=pre_process.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_transform,y,test_size=.10,random_state=101)
```

```python
x_train
```

```python
y_train
```

# **Logistic Regression**

```python
# Import model
from sklearn.linear_model import LogisticRegression

print('Logistic Regression')
# Create instance of model
log_reg = LogisticRegression()

# Pass training data into model
log_reg.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
# prediction from the model
y_pred_log_reg = log_reg.predict(x_test)
# Score It

print('Logistic Regression')
# Accuracy
print('--'*30)
log_reg_accuracy = round(accuracy_score(y_test, y_pred_log_reg) * 100,2)
print('Accuracy', log_reg_accuracy,'%')
```

```python
from sklearn.metrics import precision_score,recall_score,confusion_matrix
precision=precision_score(y_test,y_pred_log_reg)
recall=recall_score(y_test,y_pred_log_reg)
print(f"precision:{precision:.2f}")
print(f"recall:{recall:.2f}")
print("-----"*20)
confusion=confusion_matrix(y_test,y_pred_log_reg)
print(confusion)
sns.heatmap(confusion,annot=True,fmt="d")
```

# **Decision tree classifier**

```python
from sklearn.tree import DecisionTreeClassifier

print('Decision Tree Classifier')
# Create instance of model
Dtree = DecisionTreeClassifier()

# Pass training data into model
Dtree.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
# prediction from the model
y_pred_Dtree = Dtree.predict(x_test)
# Score It

print('Decision Tree Classifier')
# Accuracy
print('--'*30)
Dtree_accuracy = round(accuracy_score(y_test, y_pred_Dtree) * 100,2)
print('Accuracy', Dtree_accuracy,'%')
```

```python
from sklearn.metrics import precision_score,recall_score,confusion_matrix
precision=precision_score(y_test,y_pred_Dtree)
recall=recall_score(y_test,y_pred_Dtree)
print(f"precision:{precision:.2f}")
print(f"recall:{recall:.2f}")
print("---"*25)
confusion=confusion_matrix(y_test,y_pred_Dtree)
print(confusion)
sns.heatmap(confusion,annot=True,fmt="d")
```

# **Random forest classifier**

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred_rfc=rfc.predict(x_test)
accuracy=round(accuracy_score(y_pred_rfc,y_test)*100,2)
print("accuracy is",accuracy,"%")
```

```python
from sklearn.metrics import precision_score,recall_score,confusion_matrix
precision=precision_score(y_test,y_pred_rfc)
recall=recall_score(y_test,y_pred_rfc)
print(f"precision:{precision:.2f}")
print(f"recall:{recall:.2f}")
print("---"*25)
confusion=confusion_matrix(y_test,y_pred_rfc)
print(confusion)
sns.heatmap(confusion,annot=True,fmt="d")
```
